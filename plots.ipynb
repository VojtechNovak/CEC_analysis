```python
import pandas as pd
import matplotlib.pyplot as plt

# ==========================================
# 1. Data & Setup
# ==========================================
data = [
    # --- 2010 (LSGO) ---
    {"Year": 2010, "Rank": 1, "Algorithm": "MA-SW-Chains", "Type": "Memetic/Hybrid"},
    {"Year": 2010, "Rank": 2, "Algorithm": "2-Stage Ensemble", "Type": "Ensemble/Hybrid"},
    {"Year": 2010, "Rank": 3, "Algorithm": "MultiPSO+Harmony", "Type": "Swarm (PSO)"},
    {"Year": 2010, "Rank": 4, "Algorithm": "DASA", "Type": "Swarm (ACO/Ant)"},
    {"Year": 2010, "Rank": 5, "Algorithm": "SADE", "Type": "Differential Evolution"},
    {"Year": 2010, "Rank": 6, "Algorithm": "DECC-D", "Type": "Differential Evolution"},
    {"Year": 2010, "Rank": 7, "Algorithm": "Locust Swarms", "Type": "Swarm (Other)"},
    {"Year": 2010, "Rank": 8, "Algorithm": "SDENS", "Type": "Differential Evolution"},
    {"Year": 2010, "Rank": 9, "Algorithm": "Classic DE", "Type": "Differential Evolution"},
    {"Year": 2010, "Rank": 10, "Algorithm": "Classic DE", "Type": "Differential Evolution"},

    # --- 2011 (Real World) ---
    {"Year": 2011, "Rank": 1, "Algorithm": "GA-MPC", "Type": "Genetic Algorithm"},
    {"Year": 2011, "Rank": 2, "Algorithm": "SAMODE", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 3, "Algorithm": "EA-DE-MA", "Type": "Memetic/Hybrid"},
    {"Year": 2011, "Rank": 4, "Algorithm": "WI-DE", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 5, "Algorithm": "Adap. DE", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 6, "Algorithm": "DE - Acr", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 7, "Algorithm": "ED-DE", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 8, "Algorithm": "DE-RHC", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 9, "Algorithm": "RGA", "Type": "Genetic Algorithm"},
    {"Year": 2011, "Rank": 10, "Algorithm": "Mod-DE-LS", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 11, "Algorithm": "msSBX-GA", "Type": "Genetic Algorithm"},
    {"Year": 2011, "Rank": 12, "Algorithm": "ENSML-DE", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 13, "Algorithm": "CDASA", "Type": "Swarm (ACO/Ant)"},

    # --- 2013 (R-P) ---
    {"Year": 2013, "Rank": 1, "Algorithm": "ICMAES-ILS", "Type": "Evolution Strategy"},
    {"Year": 2013, "Rank": 2, "Algorithm": "NBIPOPaCMA", "Type": "Evolution Strategy"},
    {"Year": 2013, "Rank": 3, "Algorithm": "DRMA", "Type": "Memetic/Hybrid"},
    {"Year": 2013, "Rank": 4, "Algorithm": "SHADE", "Type": "Differential Evolution"},
    {"Year": 2013, "Rank": 5, "Algorithm": "LS-MPC", "Type": "EDA/GA"},

    # --- 2014 ---
    {"Year": 2014, "Rank": 1, "Algorithm": "L-SHADE", "Type": "Differential Evolution"},
    {"Year": 2014, "Rank": 2, "Algorithm": "GaAPADE", "Type": "Differential Evolution"},
    {"Year": 2014, "Rank": 3, "Algorithm": "MVMO-SH", "Type": "Other"},
    {"Year": 2014, "Rank": 4, "Algorithm": "OptBees", "Type": "Swarm (Other)"},
    {"Year": 2014, "Rank": 5, "Algorithm": "RM-MEDA", "Type": "EDA/GA"},
    {"Year": 2014, "Rank": 6, "Algorithm": "UMOEAs", "Type": "Memetic/Hybrid"},

    # --- 2015 ---
    {"Year": 2015, "Rank": 1, "Algorithm": "SPS-L-SHADE-EIG", "Type": "Differential Evolution"},
    {"Year": 2015, "Rank": 2, "Algorithm": "DEsPA", "Type": "Differential Evolution"},
    {"Year": 2015, "Rank": 3, "Algorithm": "LSHADE-ND", "Type": "Differential Evolution"},
    {"Year": 2015, "Rank": 4, "Algorithm": "L-SHADE", "Type": "Differential Evolution"},
    {"Year": 2015, "Rank": 5, "Algorithm": "MVMO-PH", "Type": "Other"},
    {"Year": 2015, "Rank": 6, "Algorithm": "DE-b6e6rl", "Type": "Differential Evolution"},
    {"Year": 2015, "Rank": 7, "Algorithm": "SP-UCI", "Type": "Differential Evolution"},

    # --- 2016 ---
    {"Year": 2016, "Rank": 1, "Algorithm": "LSHADE-EpSin", "Type": "Differential Evolution"},
    {"Year": 2016, "Rank": 2, "Algorithm": "UMOEA-II", "Type": "Memetic/Hybrid"},
    {"Year": 2016, "Rank": 3, "Algorithm": "iL-SHADE", "Type": "Differential Evolution"},
    {"Year": 2016, "Rank": 4, "Algorithm": "LSHADE44", "Type": "Differential Evolution"},
    {"Year": 2016, "Rank": 5, "Algorithm": "L-SHADE", "Type": "Differential Evolution"},
    {"Year": 2016, "Rank": 6, "Algorithm": "ECL-L-SHADE", "Type": "Differential Evolution"},
    {"Year": 2016, "Rank": 7, "Algorithm": "HGO-LSHADE", "Type": "Differential Evolution"},

    # --- 2017 ---
    {"Year": 2017, "Rank": 1, "Algorithm": "CAL-SHADE", "Type": "Differential Evolution"},
    {"Year": 2017, "Rank": 2, "Algorithm": "LSHADE44+IDE", "Type": "Differential Evolution"},
    {"Year": 2017, "Rank": 3, "Algorithm": "UDE", "Type": "Differential Evolution"},
    {"Year": 2017, "Rank": 4, "Algorithm": "LSHADE44", "Type": "Differential Evolution"},
    {"Year": 2017, "Rank": 5, "Algorithm": "MA-ES", "Type": "Evolution Strategy"},
    {"Year": 2017, "Rank": 6, "Algorithm": "LSHADE-IEpsilon", "Type": "Differential Evolution"},
    {"Year": 2017, "Rank": 7, "Algorithm": "IUDE", "Type": "Differential Evolution"},
    {"Year": 2017, "Rank": 8, "Algorithm": "L-SHADE+BOC", "Type": "Differential Evolution"},
    {"Year": 2017, "Rank": 9, "Algorithm": "HECO-DE", "Type": "Differential Evolution"},

    # --- 2018 ---
    {"Year": 2018, "Rank": 1, "Algorithm": "HS-ES", "Type": "Evolution Strategy"},
    {"Year": 2018, "Rank": 2, "Algorithm": "LSHADE-ESP", "Type": "Differential Evolution"},
    {"Year": 2018, "Rank": 3, "Algorithm": "ELSHADE-SPACMA", "Type": "Differential Evolution"},
    {"Year": 2018, "Rank": 4, "Algorithm": "EBOwithCMAR", "Type": "Hybrid"}, # Butterfly + CMA
    {"Year": 2018, "Rank": 5, "Algorithm": "UMOEAsII", "Type": "Memetic/Hybrid"},
    {"Year": 2018, "Rank": 6, "Algorithm": "MVMO-PH", "Type": "Other"},

    # --- 2019 (100-Digit) ---
    {"Year": 2019, "Rank": 1, "Algorithm": "jDE100", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 2, "Algorithm": "DISHchain1e+12", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 3, "Algorithm": "HyDE-DF", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 4, "Algorithm": "SOMA T3A", "Type": "Swarm (Other)"},
    {"Year": 2019, "Rank": 5, "Algorithm": "LSHADE-USM", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 6, "Algorithm": "SOMA Pareto", "Type": "Swarm (Other)"},
    {"Year": 2019, "Rank": 7, "Algorithm": "rCIPDE", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 8, "Algorithm": "Co-Op", "Type": "Cooperative"},
    {"Year": 2019, "Rank": 9, "Algorithm": "DISH", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 10, "Algorithm": "jDE", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 11, "Algorithm": "mL-SHADE", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 12, "Algorithm": "GADE", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 13, "Algorithm": "CMEAL", "Type": "Cooperative"},
    {"Year": 2019, "Rank": 14, "Algorithm": "HTPC", "Type": "Hybrid"}, # PSO + CMA
    {"Year": 2019, "Rank": 15, "Algorithm": "UMDE-MS", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 16, "Algorithm": "DLABC", "Type": "Swarm (Other)"}, # Artificial Bee Colony
    {"Year": 2019, "Rank": 17, "Algorithm": "MiLSHADE-LSP", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 18, "Algorithm": "ESP-SOMA", "Type": "Swarm (Other)"},

    # --- 2020 ---
    {"Year": 2020, "Rank": 1, "Algorithm": "IMODE", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 2, "Algorithm": "AGSK", "Type": "Differential Evolution"}, # Societal/Knowledge Sharing
    {"Year": 2020, "Rank": 3, "Algorithm": "j2020", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 4, "Algorithm": "OLSHADE", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 5, "Algorithm": "jDE100e", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 6, "Algorithm": "RASP-SHADE", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 7, "Algorithm": "mpmL-SHADE", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 8, "Algorithm": "SOMA-CL", "Type": "Swarm (Other)"},
    {"Year": 2020, "Rank": 9, "Algorithm": "MP-EEH", "Type": "Hybrid"},
    {"Year": 2020, "Rank": 10, "Algorithm": "DISH-XX", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 11, "Algorithm": "CSsin", "Type": "Swarm (Other)"}, # Cuckoo Search

    # --- 2021 ---
    {"Year": 2021, "Rank": 1, "Algorithm": "NL-SHADE-RSP", "Type": "Differential Evolution"},
    {"Year": 2021, "Rank": 2, "Algorithm": "jDE21", "Type": "Differential Evolution"},
    {"Year": 2021, "Rank": 3, "Algorithm": "APGSK-IMODE", "Type": "Hybrid"}, # GSK + DE
    {"Year": 2021, "Rank": 4, "Algorithm": "DEDMNA", "Type": "Differential Evolution"},
    {"Year": 2021, "Rank": 5, "Algorithm": "MadDE", "Type": "Differential Evolution"},
    {"Year": 2021, "Rank": 6, "Algorithm": "MLS-LSHADE", "Type": "Differential Evolution"},
    {"Year": 2021, "Rank": 7, "Algorithm": "LSHADE", "Type": "Differential Evolution"},
    {"Year": 2021, "Rank": 8, "Algorithm": "SOMA-CLP", "Type": "Swarm (Other)"},
    {"Year": 2021, "Rank": 9, "Algorithm": "RB-IPOP-CMAES", "Type": "Evolution Strategy"},

    # --- 2022 ---
    {"Year": 2022, "Rank": 1, "Algorithm": "EA4eigN100", "Type": "Cooperative"},
    {"Year": 2022, "Rank": 2, "Algorithm": "NL-SHADE-LBC", "Type": "Differential Evolution"},
    {"Year": 2022, "Rank": 3, "Algorithm": "NL-SHADE-RSP-MID", "Type": "Differential Evolution"},
    {"Year": 2022, "Rank": 4, "Algorithm": "S-LSHADE-DP", "Type": "Differential Evolution"},
    {"Year": 2022, "Rank": 5, "Algorithm": "jSObinexpEig", "Type": "Differential Evolution"},
    {"Year": 2022, "Rank": 6, "Algorithm": "MTT-SHADE", "Type": "Differential Evolution"},
    {"Year": 2022, "Rank": 7, "Algorithm": "IUMOEAII", "Type": "Memetic/Hybrid"},
    {"Year": 2022, "Rank": 8, "Algorithm": "IMPML-SHADE", "Type": "Differential Evolution"},
    {"Year": 2022, "Rank": 9, "Algorithm": "NLSOMACLP", "Type": "Swarm (Other)"},
    {"Year": 2022, "Rank": 10, "Algorithm": "ZOCMAES", "Type": "Evolution Strategy"},
    {"Year": 2022, "Rank": 11, "Algorithm": "OMCSOMA", "Type": "Swarm (Other)"},
    {"Year": 2022, "Rank": 12, "Algorithm": "Co-PPSO", "Type": "Swarm (PSO)"},
    {"Year": 2022, "Rank": 13, "Algorithm": "SPHI1-Ensemble", "Type": "Ensemble/Hybrid"},

    # --- 2024 ---
    {"Year": 2024, "Rank": 1, "Algorithm": "L-SRDE", "Type": "Differential Evolution"},
    {"Year": 2024, "Rank": 2, "Algorithm": "RDE", "Type": "Differential Evolution"},
    {"Year": 2024, "Rank": 3, "Algorithm": "BlockEA", "Type": "Hybrid"}, # EA with block strategies
    {"Year": 2024, "Rank": 4, "Algorithm": "mLSHADE", "Type": "Differential Evolution"},
    {"Year": 2024, "Rank": 5, "Algorithm": "jSO", "Type": "Differential Evolution"},
    {"Year": 2024, "Rank": 6, "Algorithm": "iEACOP", "Type": "Hybrid"},
]

df = pd.DataFrame(data)

def simplify_cat(t):
    if "Differential Evolution" in t: return "DE"
    if "Evolution Strategy" in t: return "ES/CMA"
    if "Swarm" in t: return "Swarm"
    if "Genetic" in t or "EDA" in t: return "GA/EDA"
    return "Hybrid/Other"

df['Category'] = df['Type'].apply(simplify_cat)

ACADEMIC_COLORS = {
    "DE": "#1f77b4", 
    "ES/CMA": "#ff7f0e", 
    "Swarm": "#2ca02c", 
    "GA/EDA": "#d62728", 
    "Hybrid/Other": "#7f7f7f"
}

# ==========================================
# 2. Plotting Logic
# ==========================================
def annotate_bars(ax, pivot_data, year_df):
    for i, year in enumerate(pivot_data.index):
        top3 = year_df[(year_df['Year'] == year) & (year_df['Rank'] <= 3)].sort_values('Rank')
        cum_sum = 0
        for category in pivot_data.columns:
            height = pivot_data.loc[year, category]
            relevant_top = top3[top3['Category'] == category]
            
            # Lowered threshold from 8 to 7 to capture 1/13th segments
            # Also ensures we don't try to plot if height is 0
            if not relevant_top.empty and height > 7: 
                label_parts = [f"{row['Rank']}. {row['Algorithm']}" for _, row in relevant_top.iterrows()]
                label_text = "\n".join(label_parts)
                y_pos = cum_sum + (height / 2)
                
                ax.text(i, y_pos, label_text, ha='center', va='center', 
                        fontsize=13, color='black')
            
            cum_sum += height

def plot_stacked_v2_split_modified():
    # Split data
    df_early = df[df['Year'] <= 2016]
    df_late = df[df['Year'] >= 2017]
    
    pivot_early = df_early.groupby(['Year', 'Category']).size().unstack(fill_value=0)
    pct_early = pivot_early.div(pivot_early.sum(axis=1), axis=0) * 100
    
    pivot_late = df_late.groupby(['Year', 'Category']).size().unstack(fill_value=0)
    pct_late = pivot_late.div(pivot_late.sum(axis=1), axis=0) * 100
    
    # Create subplots (2 rows, 1 column), slightly taller to accommodate larger text
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12), sharey=True)
    
    # --- Plot Early ---
    pct_early.plot(kind='bar', stacked=True, ax=ax1, width=0.85, legend=False,
                   color=[ACADEMIC_COLORS.get(x, '#333') for x in pct_early.columns])
    annotate_bars(ax1, pct_early, df_early)
    
    # Larger Titles and Labels
    #ax1.set_title("Part 1: CEC 2010 - 2016", fontsize=16, pad=10, fontweight='bold')
    ax1.set_ylabel("Percentage (%)", fontsize=14)
    ax1.set_xlabel("")
    ax1.tick_params(axis='x', rotation=0, labelsize=14)
    ax1.tick_params(axis='y', labelsize=12)
    
    # --- Plot Late ---
    pct_late.plot(kind='bar', stacked=True, ax=ax2, width=0.85, legend=False,
                  color=[ACADEMIC_COLORS.get(x, '#333') for x in pct_late.columns])
    annotate_bars(ax2, pct_late, df_late)
    
    #ax2.set_title("Part 2: CEC 2017 - 2024", fontsize=16, pad=10, fontweight='bold')
    ax2.set_ylabel("Percentage (%)", fontsize=14)
    ax2.set_xlabel(" ", fontsize=14, labelpad=10)
    ax2.tick_params(axis='x', rotation=0, labelsize=14)
    ax2.tick_params(axis='y', labelsize=12)

    # --- Layout & Legend ---
    # Retrieve handles and labels for legend
    handles, labels = ax2.get_legend_handles_labels()
    
    # Legend: Larger font (14), reduced vertical space via bbox_to_anchor
    # Using 'fig.legend' allows us to place it relative to the figure bottom
    fig.legend(handles, labels, loc='lower center', 
               bbox_to_anchor=(0.5, 0.02), 
               ncol=5, title="Algorithm Family", 
               fontsize=14, title_fontsize=14)
    
    # Tight adjustment: hspace=0.05 (minimal vertical gap between plots)
    # bottom=0.12 (leaves just enough room for the legend)
    plt.subplots_adjust(hspace=0.08, bottom=0.12, top=0.95)
    
    filename = "stacked_bar_v2_final.pdf"
    plt.savefig(filename, format='pdf', bbox_inches='tight')
    print(f"Saved {filename}")
    plt.show()

if __name__ == "__main__":
    plot_stacked_v2_split_modified()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# ... [Keep your data and simplify_cat function exactly the same] ...

def plot_stacked_v2_split_modified():
    # Split data
    df_early = df[df['Year'] <= 2016]
    df_late = df[df['Year'] >= 2017]
    
    # Define order to ensure consistency across plots
    category_order = ["DE", "ES/CMA", "GA/EDA", "Swarm", "Hybrid/Other"]
    
    pivot_early = df_early.groupby(['Year', 'Category']).size().unstack(fill_value=0)
    pivot_early = pivot_early.reindex(columns=category_order, fill_value=0)
    pct_early = pivot_early.div(pivot_early.sum(axis=1), axis=0) * 100
    
    pivot_late = df_late.groupby(['Year', 'Category']).size().unstack(fill_value=0)
    pivot_late = pivot_late.reindex(columns=category_order, fill_value=0)
    pct_late = pivot_late.div(pivot_late.sum(axis=1), axis=0) * 100
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12), sharey=True)
    
    # Colors mapped to the explicit category order
    colors = [ACADEMIC_COLORS[cat] for cat in category_order]
    
    # --- Plot Early ---
    pct_early.plot(kind='bar', stacked=True, ax=ax1, width=0.85, legend=False, color=colors)
    annotate_bars(ax1, pct_early, df_early)
    ax1.set_ylabel("Percentage (%)", fontsize=14)
    ax1.tick_params(axis='x', rotation=0, labelsize=14)
    
    # --- Plot Late ---
    pct_late.plot(kind='bar', stacked=True, ax=ax2, width=0.85, legend=False, color=colors)
    annotate_bars(ax2, pct_late, df_late)
    ax2.set_ylabel("Percentage (%)", fontsize=14)
    ax2.tick_params(axis='x', rotation=0, labelsize=14)

    # --- FIXED LEGEND LOGIC ---
    # Create manual legend handles to ensure "GA/EDA" (Red) is always included
    legend_handles = [
        mpatches.Patch(color=ACADEMIC_COLORS[cat], label=cat) 
        for cat in category_order
    ]
    
    fig.legend(handles=legend_handles, loc='lower center', 
               bbox_to_anchor=(0.5, 0.02), 
               ncol=5, title="Algorithm Family", 
               fontsize=14, title_fontsize=14)
    
    plt.subplots_adjust(hspace=0.08, bottom=0.12, top=0.95)
    filename = "stacked_bar_v2_final.pdf"
    plt.savefig(filename, format='pdf', bbox_inches='tight')
    print(f"Saved {filename}")
    plt.show()

if __name__ == "__main__":
    plot_stacked_v2_split_modified()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# ---------------------------------------------------------
# 1. THE DATASET
# ---------------------------------------------------------
data = [
    # --- 2010 (LSGO) ---
    {"Year": 2010, "Rank": 1, "Algorithm": "MA-SW-Chains", "Type": "Memetic/Hybrid"},
    {"Year": 2010, "Rank": 2, "Algorithm": "2-Stage Ensemble", "Type": "Ensemble/Hybrid"},
    {"Year": 2010, "Rank": 3, "Algorithm": "MultiPSO+Harmony", "Type": "Swarm (PSO)"},
    {"Year": 2010, "Rank": 4, "Algorithm": "DASA", "Type": "Swarm (ACO/Ant)"},
    {"Year": 2010, "Rank": 5, "Algorithm": "SADE", "Type": "Differential Evolution"},
    {"Year": 2010, "Rank": 6, "Algorithm": "DECC-D", "Type": "Differential Evolution"},
    {"Year": 2010, "Rank": 7, "Algorithm": "Locust Swarms", "Type": "Swarm (Other)"},
    {"Year": 2010, "Rank": 8, "Algorithm": "SDENS", "Type": "Differential Evolution"},

    # --- 2011 (Real World) ---
    {"Year": 2011, "Rank": 1, "Algorithm": "GA-MPC", "Type": "Genetic Algorithm"},
    {"Year": 2011, "Rank": 2, "Algorithm": "DE - Acr", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 3, "Algorithm": "SAMODE", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 4, "Algorithm": "Adap.DE", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 5, "Algorithm": "ED-DE", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 6, "Algorithm": "OXCoDE", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 7, "Algorithm": "ENSML-DE", "Type": "Differential Evolution"},
    {"Year": 2011, "Rank": 8, "Algorithm": "WL-DE", "Type": "Differential Evolution"},

    # --- 2013 (RP) ---
    {"Year": 2013, "Rank": 1, "Algorithm": "ICMAES-ILS", "Type": "Evolution Strategy"},
    {"Year": 2013, "Rank": 2, "Algorithm": "NBIPOPaCMA", "Type": "Evolution Strategy"},
    {"Year": 2013, "Rank": 3, "Algorithm": "DRMA", "Type": "Memetic/Hybrid"},
    {"Year": 2013, "Rank": 4, "Algorithm": "SHADE", "Type": "Differential Evolution"},
    {"Year": 2013, "Rank": 5, "Algorithm": "LS-MPC", "Type": "EDA/GA"},

    # --- 2014 ---
    {"Year": 2014, "Rank": 1, "Algorithm": "L-SHADE", "Type": "Differential Evolution"},
    {"Year": 2014, "Rank": 2, "Algorithm": "GaAPADE", "Type": "Differential Evolution"},
    {"Year": 2014, "Rank": 3, "Algorithm": "MVMO-SH", "Type": "Other"},
    {"Year": 2014, "Rank": 4, "Algorithm": "OptBees", "Type": "Swarm (Other)"},
    {"Year": 2014, "Rank": 5, "Algorithm": "RM-MEDA", "Type": "EDA/GA"},
    {"Year": 2014, "Rank": 6, "Algorithm": "UMOEAs", "Type": "Memetic/Hybrid"},

    # --- 2015 ---
    {"Year": 2015, "Rank": 1, "Algorithm": "SPS-L-SHADE-EIG", "Type": "Differential Evolution"},
    {"Year": 2015, "Rank": 2, "Algorithm": "DEsPA", "Type": "Differential Evolution"},
    {"Year": 2015, "Rank": 3, "Algorithm": "LSHADE-ND", "Type": "Differential Evolution"},
    {"Year": 2015, "Rank": 4, "Algorithm": "L-SHADE", "Type": "Differential Evolution"},
    {"Year": 2015, "Rank": 5, "Algorithm": "MVMO-PH", "Type": "Other"},
    {"Year": 2015, "Rank": 6, "Algorithm": "DE-b6e6rl", "Type": "Differential Evolution"},
    {"Year": 2015, "Rank": 7, "Algorithm": "SP-UCI", "Type": "Differential Evolution"},

    # --- 2016 ---
    {"Year": 2016, "Rank": 1, "Algorithm": "LSHADE-EpSin", "Type": "Differential Evolution"},
    {"Year": 2016, "Rank": 2, "Algorithm": "UMOEA-II", "Type": "Memetic/Hybrid"},
    {"Year": 2016, "Rank": 3, "Algorithm": "iL-SHADE", "Type": "Differential Evolution"},
    {"Year": 2016, "Rank": 4, "Algorithm": "LSHADE44", "Type": "Differential Evolution"},
    {"Year": 2016, "Rank": 5, "Algorithm": "L-SHADE", "Type": "Differential Evolution"},
    {"Year": 2016, "Rank": 6, "Algorithm": "ECL-L-SHADE", "Type": "Differential Evolution"},
    {"Year": 2016, "Rank": 7, "Algorithm": "HGO-LSHADE", "Type": "Differential Evolution"},

    # --- 2017 ---
    {"Year": 2017, "Rank": 1, "Algorithm": "CAL-SHADE", "Type": "Differential Evolution"},
    {"Year": 2017, "Rank": 2, "Algorithm": "LSHADE44+IDE", "Type": "Differential Evolution"},
    {"Year": 2017, "Rank": 3, "Algorithm": "UDE", "Type": "Differential Evolution"},
    {"Year": 2017, "Rank": 4, "Algorithm": "LSHADE44", "Type": "Differential Evolution"},
    {"Year": 2017, "Rank": 5, "Algorithm": "MA-ES", "Type": "Evolution Strategy"},
    {"Year": 2017, "Rank": 6, "Algorithm": "LSHADE-IEpsilon", "Type": "Differential Evolution"},
    {"Year": 2017, "Rank": 7, "Algorithm": "IUDE", "Type": "Differential Evolution"},
    {"Year": 2017, "Rank": 8, "Algorithm": "L-SHADE+BOC", "Type": "Differential Evolution"},

    # --- 2018 ---
    {"Year": 2018, "Rank": 1, "Algorithm": "HS-ES", "Type": "Evolution Strategy"},
    {"Year": 2018, "Rank": 2, "Algorithm": "LSHADE-ESP", "Type": "Differential Evolution"},
    {"Year": 2018, "Rank": 3, "Algorithm": "ELSHADE-SPACMA", "Type": "Differential Evolution"},
    {"Year": 2018, "Rank": 4, "Algorithm": "EBOwithCMAR", "Type": "Hybrid"},
    {"Year": 2018, "Rank": 5, "Algorithm": "UMOEAsII", "Type": "Memetic/Hybrid"},
    {"Year": 2018, "Rank": 6, "Algorithm": "MVMO-PH", "Type": "Other"},

    # --- 2019 (100-Digit) ---
    {"Year": 2019, "Rank": 1, "Algorithm": "jDE100", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 2, "Algorithm": "DISHchain1e+12", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 3, "Algorithm": "HyDE-DF", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 4, "Algorithm": "SOMA T3A", "Type": "Swarm (Other)"},
    {"Year": 2019, "Rank": 5, "Algorithm": "LSHADE-USM", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 6, "Algorithm": "SOMA Pareto", "Type": "Swarm (Other)"},
    {"Year": 2019, "Rank": 7, "Algorithm": "rCIPDE", "Type": "Differential Evolution"},
    {"Year": 2019, "Rank": 8, "Algorithm": "Co-Op", "Type": "Cooperative"},

    # --- 2020 ---
    {"Year": 2020, "Rank": 1, "Algorithm": "IMODE", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 2, "Algorithm": "AGSK", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 3, "Algorithm": "j2020", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 4, "Algorithm": "OLSHADE", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 5, "Algorithm": "jDE100e", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 6, "Algorithm": "RASP-SHADE", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 7, "Algorithm": "mpmL-SHADE", "Type": "Differential Evolution"},
    {"Year": 2020, "Rank": 8, "Algorithm": "SOMA-CL", "Type": "Swarm (Other)"},

    # --- 2021 ---
    {"Year": 2021, "Rank": 1, "Algorithm": "NL-SHADE-RSP", "Type": "Differential Evolution"},
    {"Year": 2021, "Rank": 2, "Algorithm": "jDE21", "Type": "Differential Evolution"},
    {"Year": 2021, "Rank": 3, "Algorithm": "APGSK-IMODE", "Type": "Hybrid"},
    {"Year": 2021, "Rank": 4, "Algorithm": "DEDMNA", "Type": "Differential Evolution"},
    {"Year": 2021, "Rank": 5, "Algorithm": "MadDE", "Type": "Differential Evolution"},
    {"Year": 2021, "Rank": 6, "Algorithm": "MLS-LSHADE", "Type": "Differential Evolution"},
    {"Year": 2021, "Rank": 7, "Algorithm": "LSHADE", "Type": "Differential Evolution"},
    {"Year": 2021, "Rank": 8, "Algorithm": "SOMA-CLP", "Type": "Swarm (Other)"},

    # --- 2022 ---
    {"Year": 2022, "Rank": 1, "Algorithm": "EA4eigN100", "Type": "Cooperative"},
    {"Year": 2022, "Rank": 2, "Algorithm": "NL-SHADE-LBC", "Type": "Differential Evolution"},
    {"Year": 2022, "Rank": 3, "Algorithm": "NL-SHADE-RSP-MID", "Type": "Differential Evolution"},
    {"Year": 2022, "Rank": 4, "Algorithm": "S-LSHADE-DP", "Type": "Differential Evolution"},
    {"Year": 2022, "Rank": 5, "Algorithm": "jSObinexpEig", "Type": "Differential Evolution"},
    {"Year": 2022, "Rank": 6, "Algorithm": "MTT-SHADE", "Type": "Differential Evolution"},
    {"Year": 2022, "Rank": 7, "Algorithm": "IUMOEAII", "Type": "Memetic/Hybrid"},
    {"Year": 2022, "Rank": 8, "Algorithm": "IMPML-SHADE", "Type": "Differential Evolution"},
    
    # --- 2024 ---
    {"Year": 2024, "Rank": 1, "Algorithm": "L-SRDE", "Type": "Differential Evolution"},
    {"Year": 2024, "Rank": 2, "Algorithm": "RDE", "Type": "Differential Evolution"},
    {"Year": 2024, "Rank": 3, "Algorithm": "BlockEA", "Type": "Hybrid"},
    {"Year": 2024, "Rank": 4, "Algorithm": "mLSHADE", "Type": "Differential Evolution"},
    {"Year": 2024, "Rank": 5, "Algorithm": "jSO", "Type": "Differential Evolution"},
    {"Year": 2024, "Rank": 6, "Algorithm": "iEACOP", "Type": "Hybrid"},
]

df = pd.DataFrame(data)

# ---------------------------------------------------------
# 2. CATEGORIZATION LOGIC
# ---------------------------------------------------------
# We group algorithms to show the "Extinction" vs "Dominance"
def get_detailed_category(row):
    algo_name = row['Algorithm'].upper()
    algo_type = row['Type']
    
    # Separate SHADE-based DE from older DE to show evolution
    if 'SHADE' in algo_name or 'L-SRDE' in algo_name or 'JSO' in algo_name:
        return 'DE (SHADE/Adaptive)'
    elif algo_type == 'Differential Evolution':
        return 'DE (Canonical/Early)'
    elif 'CMA' in algo_name or 'Evolution Strategy' in algo_type:
        return 'ES/CMA'
    elif 'Swarm' in algo_type or 'PSO' in algo_name or 'ACO' in algo_type or 'AGSK' in algo_name:
        return 'Swarm'
    elif 'Genetic' in algo_type or 'EDA' in algo_type:
        return 'GA/EDA'
    else:
        # Hybrids often end up here
        return 'Hybrid/Other'

df['DetailedCategory'] = df.apply(get_detailed_category, axis=1)

# Define Colors for the ribbons
colors = {
    'DE (SHADE/Adaptive)': "#0919aa",  # Strong Blue (The Winner)
    'DE (Canonical/Early)': "#3671df", # Light Blue (The Ancestor)
    'ES/CMA': '#ff7f0e',               # Orange (The Survivor)
    'Swarm': '#2ca02c',                # Green (The Extinct)
    'GA/EDA': '#d62728',               # Red (The Extinct)
    'Hybrid/Other': '#7f7f7f'          # Grey (The Complex)
}

# The stacking order matters for visual clarity
# We put the "Extinct" ones at the bottom/top and the Dominant in the middle or bottom
category_order = ['GA/EDA', 'Swarm', 'Hybrid/Other', 'ES/CMA', 'DE (Canonical/Early)', 'DE (SHADE/Adaptive)']

# ---------------------------------------------------------
# 3. DATA PROCESSING FOR ALLUVIAL PLOT
# ---------------------------------------------------------
years = sorted(df['Year'].unique())
counts = df.groupby(['Year', 'DetailedCategory']).size().unstack(fill_value=0)
counts = counts.reindex(columns=category_order, fill_value=0)

# Normalize to percentage (0-100)
props = counts.div(counts.sum(axis=1), axis=0) * 100

# ---------------------------------------------------------
# 4. PLOTTING WITH SMOOTH CURVES (SIGMOID)
# ---------------------------------------------------------
fig, ax = plt.subplots(figsize=(18, 9))

# X-axis mapping (discrete steps)
x_map = {y: i for i, y in enumerate(years)}
x_ticks = np.arange(len(years))

# Bottom tracker
bottom_y = np.zeros(len(years))

# Helper: Sigmoid interpolation function
def sigmoid_interp(x, y1, y2):
    # Map x from [0,1] to smooth S-curve
    mu = (1 - np.cos(x * np.pi)) / 2  # Cosine interpolation is smoother than simple polynomial
    return y1 * (1 - mu) + y2 * mu

# Draw Ribbons
for cat in category_order:
    if cat not in props.columns: continue
    
    heights = props[cat].values
    top_y = bottom_y + heights
    
    # We draw segment by segment (Year i to Year i+1)
    for i in range(len(years) - 1):
        x_start = i
        x_end = i + 1
        
        # Create dense x points for smoothness
        xs = np.linspace(x_start, x_end, 100)
        norm_xs = np.linspace(0, 1, 100) # 0 to 1 for interpolation
        
        # Interpolate Bottom edge
        y_bot = sigmoid_interp(norm_xs, bottom_y[i], bottom_y[i+1])
        
        # Interpolate Top edge
        y_top = sigmoid_interp(norm_xs, top_y[i], top_y[i+1])
        
        # Fill
        ax.fill_between(xs, y_bot, y_top, color=colors[cat], alpha=0.9, edgecolor='white', linewidth=0.5)
    
    # Update bottom for next category
    bottom_y = top_y

# ---------------------------------------------------------
# 5. LABELS & ANNOTATIONS
# ---------------------------------------------------------
# Add "Winner" boxes
winners = df[df['Rank'] == 1]

for _, row in winners.iterrows():
    year = row['Year']
    algo = row['Algorithm']
    cat = row['DetailedCategory']
    
    x_idx = x_map[year]
    
    # Calculate vertical center of the ribbon at this year
    # Re-calculate cumsum up to this category
    cats_before = category_order[:category_order.index(cat)]
    y_base = 0
    for c in cats_before:
        if c in props.columns:
            y_base += props.loc[year, c]
            
    y_h = props.loc[year, cat]
    y_center = y_base + y_h / 2
    
    # Text Box
    # Adjust font color for visibility
    text_color = 'white' if cat in ['DE (SHADE/Adaptive)', 'GA/EDA', 'Swarm', 'Hybrid/Other'] else 'black'
    
    ax.text(x_idx, y_center, algo, ha='center', va='center', fontsize=9, fontweight='bold', color=text_color,
            bbox=dict(boxstyle="round,pad=0.3", fc=colors[cat], ec="black", lw=1, alpha=1.0))

# Axis formatting
ax.set_xticks(x_ticks)
ax.set_xticklabels(years, fontsize=12, fontweight='bold')
ax.set_yticks([])
ax.set_xlim(-0.5, len(years) - 0.5)
ax.set_ylim(0, 100)

# Title and Labels
plt.title("Phylogenetic Evolution of CEC Optimization Winners (2010–2024)", fontsize=18, fontweight='bold', pad=20)
plt.ylabel("Percentage of Top Algorithms (%)", fontsize=14)

# Custom Legend
patches = [mpatches.Patch(color=colors[c], label=c) for c in reversed(category_order)] # Reversed to match visual stack
plt.legend(handles=patches, loc='center left', bbox_to_anchor=(1.01, 0.5), fontsize=12, title="Algorithm Family", frameon=False)

# Clean layout
plt.tight_layout()
plt.subplots_adjust(right=0.85) # Make room for legend

# Save or Show
# plt.savefig('cec_phylogeny_stripes.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def parse_scopus_csv(file_name):
    """Parses Scopus 'Documents by year' CSV files."""
    data = []
    with open(file_name, 'r') as f:
        found_header = False
        for line in f:
            line = line.strip()
            if not found_header:
                if line.startswith('"YEAR"'):
                    found_header = True
                continue
            if not line:
                continue
            parts = line.split(',')
            if len(parts) >= 2:
                year = parts[0].strip('"')
                count = parts[1].strip('"')
                if year.isdigit() and count.isdigit():
                    data.append({'Year': int(year), 'Count': int(count)})
    return pd.DataFrame(data).sort_values('Year')

# 1. Load the three datasets
df_intersect = parse_scopus_csv('intersection-quantum-optimization.csv').rename(columns={'Count': 'Intersection'})
df_quantum = parse_scopus_csv('qcORqa.csv').rename(columns={'Count': 'Quantum'})
df_optim = parse_scopus_csv('optimORevolORmetaheur.csv').rename(columns={'Count': 'Optimization'})

# 2. Combine datasets into one table
df_combined = df_optim.merge(df_quantum, on='Year', how='outer') \
                      .merge(df_intersect, on='Year', how='outer') \
                      .fillna(0).sort_values('Year')

# 3. Filter for the 1990-2024 range
df_plot = df_combined[(df_combined['Year'] >= 1990) & (df_combined['Year'] <= 2024)]

# 4. Create the Plot
plt.figure(figsize=(8, 6))

# Plotting the lines
plt.plot(df_plot['Year'], df_plot['Optimization'], marker='o', label='Optimization OR Evolutionary OR Metaheuristic', color='#1f77b4', linewidth=2)
plt.plot(df_plot['Year'], df_plot['Quantum'], marker='o', label='Quantum Computing OR Quantum Algorithm', color='#ff7f0e', linewidth=2)
plt.plot(df_plot['Year'], df_plot['Intersection'], marker='o', label='Quantum AND Optimization (Intersection)', color='#2ca02c', linewidth=2)

# Styling
plt.title('Number of Publications Based on Search Queries (1990-2024)', fontsize=14, pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Publications', fontsize=12)
plt.legend(title='Search Queries', loc='upper left', frameon=True)
plt.grid(axis='y', linestyle='-', alpha=0.7)

# Format X-axis labels (rotate for readability)
plt.xticks(df_plot['Year'][::2], rotation=45) 

plt.tight_layout()
plt.savefig('scopus_trends_plot.pdf')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ==========================================
# 1. Configuration Toggle
# ==========================================
# Set to True for grayscale with hatching, False for academic blues/colors
USE_BW_HATCHING = False 

# ==========================================
# 2. Reclassified Data & Setup
# ==========================================
data = [
    # --- 2010 (LSGO) ---
    {"Year": 2010, "Rank": 1, "Algorithm": "MA-SW-Chains", "Category": "Hybrid/Other"},
    {"Year": 2010, "Rank": 2, "Algorithm": "2-Stage Ensemble", "Category": "Hybrid/Other"},
    {"Year": 2010, "Rank": 3, "Algorithm": "MultiPSO+Harmony", "Category": "Swarm"},
    {"Year": 2010, "Rank": 4, "Algorithm": "DASA", "Category": "Swarm"},
    {"Year": 2010, "Rank": 5, "Algorithm": "SADE", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2010, "Rank": 6, "Algorithm": "DECC-D", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2010, "Rank": 7, "Algorithm": "Locust Swarms", "Category": "Swarm"},
    {"Year": 2010, "Rank": 8, "Algorithm": "SDENS", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2010, "Rank": 9, "Algorithm": "Classic DE", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2010, "Rank": 10, "Algorithm": "Classic DE", "Category": "DE - Canonical/Adaptive"},

    # --- 2011 (Real World) ---
    {"Year": 2011, "Rank": 1, "Algorithm": "GA-MPC", "Category": "GA/EDA"},
    {"Year": 2011, "Rank": 2, "Algorithm": "SAMODE", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2011, "Rank": 3, "Algorithm": "EA-DE-MA", "Category": "Hybrid/Other"},
    {"Year": 2011, "Rank": 4, "Algorithm": "WI-DE", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2011, "Rank": 5, "Algorithm": "Adap. DE", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2011, "Rank": 6, "Algorithm": "DE - Acr", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2011, "Rank": 7, "Algorithm": "ED-DE", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2011, "Rank": 8, "Algorithm": "DE-RHC", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2011, "Rank": 9, "Algorithm": "RGA", "Category": "GA/EDA"},
    {"Year": 2011, "Rank": 10, "Algorithm": "Mod-DE-LS", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2011, "Rank": 11, "Algorithm": "msSBX-GA", "Category": "GA/EDA"},
    {"Year": 2011, "Rank": 12, "Algorithm": "ENSML-DE", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2011, "Rank": 13, "Algorithm": "CDASA", "Category": "Swarm"},

    # --- 2013 (R-P) ---
    {"Year": 2013, "Rank": 1, "Algorithm": "ICMAES-ILS", "Category": "ES/CMA"},
    {"Year": 2013, "Rank": 2, "Algorithm": "NBIPOPaCMA", "Category": "ES/CMA"},
    {"Year": 2013, "Rank": 3, "Algorithm": "DRMA", "Category": "Hybrid/Other"},
    {"Year": 2013, "Rank": 4, "Algorithm": "SHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2013, "Rank": 5, "Algorithm": "LS-MPC", "Category": "GA/EDA"},

    # --- 2014 ---
    {"Year": 2014, "Rank": 1, "Algorithm": "L-SHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2014, "Rank": 2, "Algorithm": "GaAPADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2014, "Rank": 3, "Algorithm": "MVMO-SH", "Category": "Hybrid/Other"},
    {"Year": 2014, "Rank": 4, "Algorithm": "OptBees", "Category": "Swarm"},
    {"Year": 2014, "Rank": 5, "Algorithm": "RM-MEDA", "Category": "GA/EDA"},
    {"Year": 2014, "Rank": 6, "Algorithm": "UMOEAs", "Category": "Hybrid/Other"},

    # --- 2015 ---
    {"Year": 2015, "Rank": 1, "Algorithm": "SPS-L-SHADE-EIG", "Category": "DE - SHADE Lineage"},
    {"Year": 2015, "Rank": 2, "Algorithm": "DEsPA", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2015, "Rank": 3, "Algorithm": "LSHADE-ND", "Category": "DE - SHADE Lineage"},
    {"Year": 2015, "Rank": 4, "Algorithm": "L-SHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2015, "Rank": 5, "Algorithm": "MVMO-PH", "Category": "Hybrid/Other"},
    {"Year": 2015, "Rank": 6, "Algorithm": "DE-b6e6rl", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2015, "Rank": 7, "Algorithm": "SP-UCI", "Category": "DE - Canonical/Adaptive"},

    # --- 2016 ---
    {"Year": 2016, "Rank": 1, "Algorithm": "LSHADE-EpSin", "Category": "DE - SHADE Lineage"},
    {"Year": 2016, "Rank": 2, "Algorithm": "UMOEA-II", "Category": "Hybrid/Other"},
    {"Year": 2016, "Rank": 3, "Algorithm": "iL-SHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2016, "Rank": 4, "Algorithm": "LSHADE44", "Category": "DE - SHADE Lineage"},
    {"Year": 2016, "Rank": 5, "Algorithm": "L-SHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2016, "Rank": 6, "Algorithm": "ECL-L-SHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2016, "Rank": 7, "Algorithm": "HGO-LSHADE", "Category": "DE - SHADE Lineage"},

    # --- 2017 ---
    {"Year": 2017, "Rank": 1, "Algorithm": "EBOwithCMAR", "Category": "ES/CMA"},
    {"Year": 2017, "Rank": 2, "Algorithm": "jSO", "Category": "DE - SHADE Lineage"},
    {"Year": 2017, "Rank": 3, "Algorithm": "LSHADE-SPACMA", "Category": "DE - SHADE Lineage"},
    {"Year": 2017, "Rank": 4, "Algorithm": "LSHADE-cnEpSin", "Category": "DE - SHADE Lineage"},
    {"Year": 2017, "Rank": 5, "Algorithm": "DES", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2017, "Rank": 6, "Algorithm": "MM-OEDA", "Category": "GA/EDA"},
    {"Year": 2017, "Rank": 7, "Algorithm": "IDEbestNsize", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2017, "Rank": 8, "Algorithm": "PPSO", "Category": "Swarm"},
    {"Year": 2017, "Rank": 9, "Algorithm": "DYYPO", "Category": "Hybrid/Other"},
    {"Year": 2017, "Rank": 10, "Algorithm": "RB-IPOP-CMA-ES", "Category": "ES/CMA"},

    # --- 2018 ---
    {"Year": 2018, "Rank": 1, "Algorithm": "HS-ES", "Category": "ES/CMA"},
    {"Year": 2018, "Rank": 2, "Algorithm": "LSHADE-ESP", "Category": "DE - SHADE Lineage"},
    {"Year": 2018, "Rank": 3, "Algorithm": "ELSHADE-SPACMA", "Category": "DE - SHADE Lineage"},
    {"Year": 2018, "Rank": 4, "Algorithm": "EBOwithCMAR", "Category": "ES/CMA"},
    {"Year": 2018, "Rank": 5, "Algorithm": "UMOEAsII", "Category": "Hybrid/Other"},
    {"Year": 2018, "Rank": 6, "Algorithm": "MVMO-PH", "Category": "Hybrid/Other"},

    # --- 2019 (100-Digit) ---
    {"Year": 2019, "Rank": 1, "Algorithm": "jDE100", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2019, "Rank": 2, "Algorithm": "DISHchain1e+12", "Category": "DE - SHADE Lineage"},
    {"Year": 2019, "Rank": 3, "Algorithm": "HyDE-DF", "Category": "DE - SHADE Lineage"},
    {"Year": 2019, "Rank": 4, "Algorithm": "SOMA T3A", "Category": "Swarm"},
    {"Year": 2019, "Rank": 5, "Algorithm": "LSHADE-USM", "Category": "DE - SHADE Lineage"},
    {"Year": 2019, "Rank": 6, "Algorithm": "SOMA Pareto", "Category": "Swarm"},
    {"Year": 2019, "Rank": 7, "Algorithm": "rCIPDE", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2019, "Rank": 8, "Algorithm": "Co-Op", "Category": "Hybrid/Other"},
    {"Year": 2019, "Rank": 9, "Algorithm": "DISH", "Category": "DE - SHADE Lineage"},
    {"Year": 2019, "Rank": 10, "Algorithm": "jDE", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2019, "Rank": 11, "Algorithm": "mL-SHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2019, "Rank": 12, "Algorithm": "GADE", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2019, "Rank": 13, "Algorithm": "CMEAL", "Category": "Hybrid/Other"},
    {"Year": 2019, "Rank": 14, "Algorithm": "HTPC", "Category": "Swarm"},
    {"Year": 2019, "Rank": 15, "Algorithm": "UMDE-MS", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2019, "Rank": 16, "Algorithm": "DLABC", "Category": "Swarm"},
    {"Year": 2019, "Rank": 17, "Algorithm": "MiLSHADE-LSP", "Category": "DE - SHADE Lineage"},
    {"Year": 2019, "Rank": 18, "Algorithm": "ESP-SOMA", "Category": "Swarm"},

    # --- 2020 ---
    {"Year": 2020, "Rank": 1, "Algorithm": "IMODE", "Category": "DE - Societal/Multi-Pop"},
    {"Year": 2020, "Rank": 2, "Algorithm": "AGSK", "Category": "DE - Societal/Multi-Pop"},
    {"Year": 2020, "Rank": 3, "Algorithm": "j2020", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2020, "Rank": 4, "Algorithm": "OLSHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2020, "Rank": 5, "Algorithm": "jDE100e", "Category": "DE - Driven Ensembles"},
    {"Year": 2020, "Rank": 6, "Algorithm": "RASP-SHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2020, "Rank": 7, "Algorithm": "mpmL-SHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2020, "Rank": 8, "Algorithm": "SOMA-CL", "Category": "Swarm"},
    {"Year": 2020, "Rank": 9, "Algorithm": "MP-EEH", "Category": "Hybrid/Other"},
    {"Year": 2020, "Rank": 10, "Algorithm": "DISH-XX", "Category": "DE - SHADE Lineage"},
    {"Year": 2020, "Rank": 11, "Algorithm": "CSsin", "Category": "Swarm"},

    # --- 2021 ---
    {"Year": 2021, "Rank": 1, "Algorithm": "NL-SHADE-RSP", "Category": "DE - SHADE Lineage"},
    {"Year": 2021, "Rank": 2, "Algorithm": "jDE21", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2021, "Rank": 3, "Algorithm": "APGSK-IMODE", "Category": "DE - Societal/Multi-Pop"},
    {"Year": 2021, "Rank": 4, "Algorithm": "DEDMNA", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2021, "Rank": 5, "Algorithm": "MadDE", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2021, "Rank": 6, "Algorithm": "MLS-LSHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2021, "Rank": 7, "Algorithm": "LSHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2021, "Rank": 8, "Algorithm": "SOMA-CLP", "Category": "Swarm"},
    {"Year": 2021, "Rank": 9, "Algorithm": "RB-IPOP-CMAES", "Category": "ES/CMA"},

    # --- 2022 ---
    {"Year": 2022, "Rank": 1, "Algorithm": "EA4eigN100", "Category": "DE - Driven Ensembles"},
    {"Year": 2022, "Rank": 2, "Algorithm": "NL-SHADE-LBC", "Category": "DE - SHADE Lineage"},
    {"Year": 2022, "Rank": 3, "Algorithm": "NL-SHADE-RSP-MID", "Category": "DE - SHADE Lineage"},
    {"Year": 2022, "Rank": 4, "Algorithm": "S-LSHADE-DP", "Category": "DE - SHADE Lineage"},
    {"Year": 2022, "Rank": 5, "Algorithm": "jSObinexpEig", "Category": "DE - Driven Ensembles"},
    {"Year": 2022, "Rank": 6, "Algorithm": "MTT-SHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2022, "Rank": 7, "Algorithm": "IUMOEAII", "Category": "Hybrid/Other"},
    {"Year": 2022, "Rank": 8, "Algorithm": "IMPML-SHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2022, "Rank": 9, "Algorithm": "NLSOMACLP", "Category": "Swarm"},
    {"Year": 2022, "Rank": 10, "Algorithm": "ZOCMAES", "Category": "ES/CMA"},
    {"Year": 2022, "Rank": 11, "Algorithm": "OMCSOMA", "Category": "Swarm"},
    {"Year": 2022, "Rank": 12, "Algorithm": "Co-PPSO", "Category": "Swarm"},
    {"Year": 2022, "Rank": 13, "Algorithm": "SPHI1-Ensemble", "Category": "Hybrid/Other"},

    # --- 2024 ---
    {"Year": 2024, "Rank": 1, "Algorithm": "L-SRTDE", "Category": "DE - SHADE Lineage"},
    {"Year": 2024, "Rank": 2, "Algorithm": "RDE", "Category": "DE - Canonical/Adaptive"},
    {"Year": 2024, "Rank": 3, "Algorithm": "BlockEA", "Category": "DE - Driven Ensembles"},
    {"Year": 2024, "Rank": 4, "Algorithm": "mLSHADE", "Category": "DE - SHADE Lineage"},
    {"Year": 2024, "Rank": 5, "Algorithm": "jSO", "Category": "DE - SHADE Lineage"},
    {"Year": 2024, "Rank": 6, "Algorithm": "iEACOP", "Category": "Hybrid/Other"},
]

df = pd.DataFrame(data)

# Category styling dictionaries
ORDERED_CATEGORIES = [
    "DE - SHADE Lineage",
    "DE - Societal/Multi-Pop",
    "DE - Driven Ensembles",
    "DE - Canonical/Adaptive",
    "ES/CMA",
    "Swarm",
    "GA/EDA",
    "Hybrid/Other"
]

COLOR_MAP = {
    "DE - SHADE Lineage": "#2831af",         # Dark Blue
    "DE - Societal/Multi-Pop": "#1462c9",    # Medium Blue
    "DE - Driven Ensembles": "#2081b9",      # Light Blue
    "DE - Canonical/Adaptive": "#5eaddf",    # Very Light Blue
    "ES/CMA": "#ff7f0e",                     # Orange
    "Swarm": "#2ca02c",                      # Green
    "GA/EDA": "#d62728",                     # Red
    "Hybrid/Other": "#7f7f7f"                # Grey
}

BW_COLORS = {
    "DE - SHADE Lineage": "#ffffff", 
    "DE - Societal/Multi-Pop": "#e0e0e0", 
    "DE - Driven Ensembles": "#b0b0b0", 
    "DE - Canonical/Adaptive": "#808080", 
    "ES/CMA": "#ffffff", 
    "Swarm": "#e0e0e0", 
    "GA/EDA": "#b0b0b0", 
    "Hybrid/Other": "#404040" 
}

HATCH_MAP = {
    "DE - SHADE Lineage": "\\\\",
    "DE - Societal/Multi-Pop": "//",
    "DE - Driven Ensembles": "xx",
    "DE - Canonical/Adaptive": "..",
    "ES/CMA": "**",
    "Swarm": "--",
    "GA/EDA": "++",
    "Hybrid/Other": "OO"
}

# ==========================================
# 3. Plotting Logic
# ==========================================
def annotate_bars(ax, pivot_data, year_df):
    for i, year in enumerate(pivot_data.index):
        top3 = year_df[(year_df['Year'] == year) & (year_df['Rank'] <= 3)].sort_values('Rank')
        cum_sum = 0
        for category in pivot_data.columns:
            height = pivot_data.loc[year, category]
            relevant_top = top3[top3['Category'] == category]
            
            if not relevant_top.empty and height > 7: 
                label_parts = [f"{row['Rank']}. {row['Algorithm']}" for _, row in relevant_top.iterrows()]
                label_text = "\n".join(label_parts)
                y_pos = cum_sum + (height / 2)
                
                ax.text(i, y_pos, label_text, ha='center', va='center', 
                        fontsize=11, color='black',
                        bbox=dict(facecolor='white', alpha=0.6, edgecolor='none', pad=1))
            
            cum_sum += height

def plot_stacked_v3():
    # Enforce column order based on our taxonomy
    df['Category'] = pd.Categorical(df['Category'], categories=ORDERED_CATEGORIES, ordered=True)
    
    df_early = df[df['Year'] <= 2016]
    df_late = df[df['Year'] >= 2017]
    
    pivot_early = df_early.groupby(['Year', 'Category'], observed=False).size().unstack(fill_value=0)
    pct_early = pivot_early.div(pivot_early.sum(axis=1), axis=0) * 100
    
    pivot_late = df_late.groupby(['Year', 'Category'], observed=False).size().unstack(fill_value=0)
    pct_late = pivot_late.div(pivot_late.sum(axis=1), axis=0) * 100
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 14), sharey=True)
    
    # Select style mappings directly from COLOR_MAP
    colors = [COLOR_MAP[x] for x in pct_early.columns]
    
    # --- Plot Early ---
    pct_early.plot(kind='bar', stacked=True, ax=ax1, width=0.85, legend=False, 
                   color=colors, edgecolor='none')
    
    # --- Plot Late ---
    pct_late.plot(kind='bar', stacked=True, ax=ax2, width=0.85, legend=False, 
                  color=colors, edgecolor='none')

    annotate_bars(ax1, pct_early, df_early)
    annotate_bars(ax2, pct_late, df_late)
    
    ax1.set_ylabel("Percentage (%)", fontsize=14)
    ax1.set_xlabel("")
    ax1.tick_params(axis='x', rotation=0, labelsize=14)
    ax1.tick_params(axis='y', labelsize=12)
    
    ax2.set_ylabel("Percentage (%)", fontsize=14)
    ax2.set_xlabel("Competition Year", fontsize=14, labelpad=10)
    ax2.tick_params(axis='x', rotation=0, labelsize=14)
    ax2.tick_params(axis='y', labelsize=12)

    handles, labels = ax2.get_legend_handles_labels()
    fig.legend(handles, labels, loc='lower center', 
               bbox_to_anchor=(0.5, 0.02), 
               ncol=4, title="Algorithm Family Framework", 
               fontsize=13, title_fontsize=14)
    
    plt.subplots_adjust(hspace=0.08, bottom=0.15, top=0.95)
    
    filename = "stacked_bar_v3_taxonomy_color.pdf"
    plt.savefig(filename, format='pdf', bbox_inches='tight')
    print(f"Saved {filename}")
    plt.show()

if __name__ == "__main__":
    plot_stacked_v3()
```
